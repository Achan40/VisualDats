```python
import pandas as pd
import numpy as np
import folium

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df_survey = pd.read_csv('C:/Users/chanm/Desktop/VisualDats/datasets/Topic_Survey_Assignment.csv')
df_survey
```

```python
# Total number of respondans per survey question
df_survey['Total'] = df_survey.sum(axis = 1)
```

```python
df_survey
```

```python
# Find proportion
# Divide each column by corresponding total, round to 4th decimal place and multiply by 100
df_survey_percent = (df_survey[['Very interested','Somewhat interested','Not interested']].div(df_survey['Total'],axis=0).round(4) * 100)
df_survey_percent['Topics'] = df_survey[['Unnamed: 0']]

# Sort by highest
df_survey_percent = df_survey_percent.sort_values(['Very interested'], ascending = False)

df_survey_percent
```

### Proportion of Respondents Interest in Data Science

```python
# Set desired colors
mycolors = ['#5cb85c','#5bc0de','#d9534f']

# Make plot
df_survey_percent.plot(kind='bar', figsize=(20, 8), width = .8, color = mycolors)

plt.title('Percentage of Respondents Interest in Data Science', size = 16) # add title to the plot
plt.yticks([])
plt.xticks([0,1,2,3,4,5],df_survey_percent['Topics'].tolist(), size = 14) # add ticks

# Remove Borders
ax = plt.gca() #gca means to get current axes, basically get the current plot instance
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

# Editing Legend
plt.legend(fontsize = 14)

# Annotate the bars with their values, Actually genius
for p in ax.patches[0:]:
    h = p.get_height()
    x = p.get_x()+p.get_width()/2.
    if h != 0:
        ax.annotate("%g" % p.get_height(), xy=(x,h), xytext=(0,4), # %g is string format
                   textcoords="offset points", ha="center", va="bottom", size = 14)
        
plt.show()
```
