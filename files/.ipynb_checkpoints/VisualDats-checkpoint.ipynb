```python
import pandas as pd
import numpy as np
import folium

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df_survey = pd.read_csv('C:/Users/chanm/Desktop/VisualDats/datasets/Topic_Survey_Assignment.csv')
df_survey
```

```python
# Total number of respondans per survey question
df_survey['Total'] = df_survey.sum(axis = 1)
```

```python
df_survey
```

```python
# Find proportion
df_survey_percent = df_survey[['Very interested','Somewhat interested','Not interested']].div(df_survey['Total'],axis=0)
# Sort by highest
df_survey_percent = df_survey_percent.sort_values(['Very interested'], ascending = False)
df_survey_percent
```

```python
# step 2: plot data
df_survey_percent.plot(kind='bar', figsize=(20, 8))

plt.xlabel('Topic') # add to x-label to the plot
plt.ylabel('Proportion') # add y-label to the plot
plt.title('Topic Interest by Proportion') # add title to the plot

plt.show()
```
